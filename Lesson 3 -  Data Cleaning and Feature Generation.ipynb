```python
import pandas as pd
pd.options.display.float_format = '{:20,.2f}'.format
import numpy as np
```

```python
data = pd.read_csv("Nashville_housing_data_2013_2016.csv")
```

```python
data.head()
```

```python
data['Average Room Area'] = data['Finished Area'] / data['Bedrooms']
```

```python
data.head()
```

```python
mean_total_price = data[~data['Total Value'].isna()]['Total Value'].mean()
data['Total Value'].fillna(mean_total_price)
```

```python
data['Average Room Area'].describe()
```

```python
data.plot.scatter(x='Average Room Area', y='Sale Price', alpha=0.05)
```
